```python
import anndata as ad
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_to_brain_sc = "/space/grp/aadrian/Pseudobulk_Function_Pipeline_HighRes/bin/preprocessing/preprocessSC_pipe/data/h5ad_datasets/processed/cpm/brain_sc_with_metadata_pc_cpm.h5ad"
brain_adata = ad.read_h5ad(path_to_brain_sc)
```

```python
brain_adata.shape
```

```python
path_to_ori_brain = "/space/grp/aadrian/Pseudobulk_Function_Pipeline_HighRes/bin/preprocessing/preprocessSC_pipe/data/h5ad_datasets/raw/brain_sc_with_metadata.h5ad"
brain_adata_ori = ad.read_h5ad(path_to_ori_brain)
brain_adata_ori.shape
```

```python
brain_adata.shape[0]/brain_adata_ori.shape[0]
```

```python
brain_adata.obs.value_counts("Cell type").reset_index()
```

```python
sns.barplot(brain_adata.obs.value_counts("Cell type").reset_index(), y = 'count', x = 'Cell type')
# Rotate x-axis tick labels by 30 degrees
plt.xticks(rotation=20)
plt.title("Counts of Cell Types")
plt.savefig("Brain_ct_counts.png", dpi = 300)
```
