```python
import pandas as pd
import anndata
import numpy as np
import seaborn as sns
```

```python
# GTEX
adata = anndata.read_h5ad('/pipeline42/datasets/Gtex/GTEx.h5ad')
```

```python
def subsetByOP(adata, OP):
    
    return adata

def group_by_op(adata):
    return adata.obs.groupby('SMTS')

def get_group_index(adata):
    
    
    grouped = group_by_op(adata)

    OPs_indexes = {}

    # WHen looping over a grouped by dataframe in pandas, first index is name of group. Second is data from that group
    for group_name, group_df in grouped:

        OPs_indexes.update({group_name: group_df.index})
            
    return OPs_indexes
    
```

```python
def sampleKOPs(adata, k = 20):
    OPs_indexes = get_group_index(adata)
    adatas = []
    
    for OP_key in OPs_indexes.keys():
        indexes = OPs_indexes[OP_key]

        if len(indexes) < k:
            continue
        
        index_samples = np.random.choice(np.array(indexes), k, replace = False)
        
        adatas.append(adata[index_samples,:])
        
    return anndata.concat(adatas)
        
        
```

```python
K=5
adata_K = sampleKOPs(adata, K)
```

```python
adata_K.write(f'/pipeline42/datasets/Gtex/GTEx_K{K}')
```
