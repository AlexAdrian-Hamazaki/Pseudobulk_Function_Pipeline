# Replacing nan with 0

### It is just this BBS1 that has nan in it

```python
BBS1 = testdf.iloc[:, -1]
```

```python
BBS1
```

```python
mylist= []

for val in BBS1:
    if pd.isna(val):
        mylist.append(0)
    else:
        mylist.append(val)
```

```python
mylist = [0 if pd.isna(val) else val for val in BBS1]
```

```python
mylist
```

```python
testdf.loc[:, 'BBS1'] = mylist
```
